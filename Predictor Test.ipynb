```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import pickle
```

```python
pd.pandas.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('processed_test.csv')
```

```python
import json
with open('sel_feat.json', 'r') as f:
    selected_feat = json.loads(f.read())['features']
```

```python
len(selected_feat)
```

```python
data
```

```python
X_test = data[selected_feat]
```

```python
X_test
```

```python
with open('house_model', 'rb') as f:
    model = pickle.load(f)
```

```python
y_pred = model.predict(X_test)
```

```python
np.exp(y_pred)
```

```python
predicted_prices = np.exp(y_pred)
```

```python
predicted_prices
```

```python
df = pd.DataFrame(predicted_prices, columns=['House_Price'])
```
