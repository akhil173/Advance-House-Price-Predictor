# Feature Selection

To select the required features to train the model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
pd.pandas.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('processed_train.csv')
```

```python
data.head()
```

```python
y_train = data[['SalePrice']]
```

```python
y_train
```

```python
X_train = data.drop(['Id','SalePrice'], axis=1)
```

```python
X_train
```

```python
### Apply Feature Selection
# first, I specify the Lasso Regression model, and I
# select a suitable alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected.

# Then I use the selectFromModel object from sklearn, which
# will select the features which coefficients are non-zero
```

```python
#Feature Selection Model
feature_sel_model = SelectFromModel(Lasso(0.005, random_state=0))
feature_sel_model.fit(X_train, y_train)
```

```python
feature_sel_model.get_params()
```

```python
feature_sel_model.get_support()
```

```python
selected_feat = X_train.columns[feature_sel_model.get_support()]
```

```python
feature_sel_model.estimator_.coef_
```

```python
selected_feat
```

```python
print('Total Features : ',X_train.shape[1])
print('Selected Features ; ',len(selected_feat))
print('Features with coefficients that shrank to 0 : ',np.sum(feature_sel_model.estimator_.coef_==0))
```

```python
X_train_final = X_train[selected_feat]
```

```python
import json
json.dumps({'features':list(selected_feat)})
```

```python
import json
with open('sel_feat.json','w') as f:
    f.write(json.dumps({'features':list(selected_feat)}))
```

```python
X_train_final
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, Y_train, Y_test = train_test_split(X_train_final, y_train, test_size=0.1)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, Y_train)
```

```python
model.score(x_test, Y_test)
```

```python
model.predict(x_test)
```

```python
np.exp(model.predict(x_test))
```
